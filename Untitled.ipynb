```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
mu = 0.2
sigma = 0.1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.lognorm.pdf(x, mu, sigma))
plt.show()
```

```python
from scipy.stats import lognorm
s = 0.999
mu, sigma = -1, 5
a, b = 0, 100
x = np.linspace(0, 100, 10000)
p = lognorm.pdf(x, s, loc=mu, scale=sigma)
plt.grid(color='r', linestyle='-', linewidth=0.1)
plt.rcParams['figure.figsize'] = [15, 15]
plt.xticks(range(0, 101, 3))
plt.plot(x, p, 'r-', lw=5, alpha=0.6, label='lognorm pdf')
```

```python
rs = lognorm.rvs(0.999, loc=-1, scale=5, size=10000)
print(list(rs)[:100])
plt.hist(rs, bins=100)
```
